# 1  
$y'=y$, con $y(0)=1$, calcula $y(1)$. El valor exacto es $e$, ¿cuàntos subintervalos necesitas para obtener tres cifras decimales de presiciòn?



Calculando la soluciòn exacta con **WxMaxima**:
![EDO1](images/EDO1.png)

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

def g(x,y):
	return y
#IMplementaciòn del mètodo de Euler
def EulerMej(xi,y,h,xf,i):
	aux=-0
	while xi<xf:
		temp=y
		y=y+h*g(xi,y)
		xi=xi+h
		i+=1   
	print("El valor aproximado en x = ", xf, " es ", "%.9f\n"% y)
	print("Iteraciones: ",i)
#Condiciones iniciales (Càlculo)
xi=0
y0=1
i=0
h=0.0025
#Iteraciones
i=0
#Valor a aproximar
xf=1
EulerMej(xi,y0,h,xf,i)
#Condiciones iniciales (Gràfica)
f=lambda x, s: np.exp(x)
t=np.arange(0, 1 + h, h)
s=np.zeros(len(t))
s[0]=y0
for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])
plt.figure(figsize = (12, 8))
plt.plot(t, s, 'b.', label='Sol. Aprox.')
plt.plot(t, np.exp(t), 'r', label='Sol. Exac.')
plt.title('Soluciòn exacta y aproximada a dy/dx=y')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

# 2  
$y'=\dfrac{4}{1+x^2}$, con $y(0)=0$, calcula $y(1)$. El valor exacto es $\pi$, ¿cuàntos subintervalos necesitas para obtener tres cifras decimales de presiciòn?

Calculando la soluciòn exacta con **WxMaxima**:
<img src="EDO2.png">

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

def g(x,y):
	return 4/(1+x*x)
#IMplementaciòn del mètodo de Euler
def EulerMej(xi,y,h,xf,i):
	aux=-0
	while xi<xf:
		temp=y
		y=y+h*g(xi,y)
		xi=xi+h
		i+=1   
	print("El valor aproximado en x = ", xf, " es ", "%.9f\n"% y)
	print("Iteraciones: ",i)
#Condiciones iniciales (Càlculo)
xi=0
y0=0
i=0
h=0.00025
#Iteraciones
i=0
#Valor a aproximar
xf=1
EulerMej(xi,y0,h,xf,i)
#Condiciones iniciales (Gràfica)
f=lambda x, s: 4/(1+(x*x))
t=np.arange(0, 1 + h, h)
s=np.zeros(len(t))
s[0]=y0
for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])
plt.figure(figsize = (12, 8))
plt.plot(t, s, 'b.--', label='Sol. Aprox.')
plt.plot(t, 4*np.arctan(t), 'r', label='Sol. Exac.')
plt.title('Soluciòn exacta y aproximada a dy/dx=y')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```

**Una falla en el mètodo de Euler** Considere la ecuaciòn  
$\dfrac{dy}{dx}=x^2+y^2,~~~~~y(0)=1$  
y usa el mètodo de Euler con los tamaños de peso: $h=0.1,0.02, 0.005$ para aproximar $y(1)$ ¿què sucede? ¿por què crees que ocurre esto?

```python
def g( x, y ):
	return (x*x+y*y)
def e1(x0,y,h1,x):
	temp = -0
	while x0 < x:
		temp = y
		y = y + h1 *g(x0, y)
		x0 = x0 + h1
	print("Con el tamaño de paso h1=", h1, "se tiene una aproximaciòn de " "%.6f"% y)

def e2(x0,y,h2,x):
	temp = -0
	while x0 < x:
		temp = y
		y = y + h2 *g(x0, y)
		x0 = x0 + h2
	print("Con el tamaño de paso h2=", h2, "se tiene una aproximaciòn de " "%.6f"% y)

def e3(x0,y,h3,x):
	temp = -0
	while x0 < x:
		temp = y
		y = y + h3 *g(x0, y)
		x0 = x0 + h3
	print("Con el tamaño de paso h3=", h3, "se tiene una aproximaciòn de " "%.6f"% y)

x0=0
y0=1
h1=0.1
h2=0.02
h3=0.005

#Valor a aproximar
x=1

e1(x0, y0, h1, x)
e2(x0, y0, h2, x)
e3(x0, y0, h3, x)
```

Podemos ver que entre màs pequeño sea el tamaño de paso, la aproximaciòn ira incrementando pero  
no se asegura que sea la soluciòn real. El método de Euler usa la línea tangente a la función al comienzo del intervalo como una estimación de la pendiente de la función sobre el intervalo, asumiendo que si el tamaño del paso es pequeño, el error será pequeño. Sabemos que la pendiente de la recta tangente es calculada por:  
  
  
$m=\dfrac{y_f-y_i}{x_f-x_i}$  
donde $xf-x_i=h$, al ser $h$ demasiado pequeño, el valor del cociente comienza a incrementarse, es obvio que al aplicar limite hacia $0$ el cociente tiende a infinito.  
Al observar el campo direccional de la ecuaciòn diferencial, para el valor de $y(1)$ se tiene que la pendiente de la soluciòn particular comienza a incrementar demasiado ràpido
